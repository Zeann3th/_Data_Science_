# Introduction to Matplotlib

### 1. Importing 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### 2. Preparing Data

```python
cities = pd.read_csv("../Databases/california_cities.csv")
cities
```

```python
# Extract Latitude & Longtitude
lat, long = cities['latd'], cities['longd']
population, area =  cities['population_total'], cities['area_total_km2']
plt.figure(figsize = (8, 6))

# Plot using Pyplot API
plt.scatter(long, lat, c = np.log10(population), s = area, cmap = 'viridis', linewidth = 0, alpha = 0.5)

# Modifications
plt.colorbar(label = 'log$_{10}$(population)')
plt.clim(3,7)
plt.axis("equal")
plt.xlabel("Longtitude")
plt.ylabel("Latitude")
plt.title("California cities: Population and Area Distribution")
plt.legend();
# Create a legend for cities size
area_range = [50, 100, 300, 500]
for area in area_range:
    plt.scatter([], [], s = area, label = str(area) + 'km$^2$', c = 'green', alpha = 0.4)
plt.legend(labelspacing = 1, title = "City Area");
```
